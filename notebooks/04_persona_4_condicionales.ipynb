# Persona 4 - Condicionales

Este notebook corresponde a la participación de la **persona 4**.

Su objetivo es aplicar `if`, `elif` y `else` para clasificar humedad, temperatura y nivel de descomposición.

````python
from pathlib import Path
import sys

# Detecta la raíz del proyecto aunque abras el notebook desde /notebooks o desde la raíz.
current = Path.cwd()
if (current / "data").exists() and (current / "src").exists():
    PROJECT_ROOT = current
elif (current.parent / "data").exists() and (current.parent / "src").exists():
    PROJECT_ROOT = current.parent
else:
    PROJECT_ROOT = current

sys.path.insert(0, str(PROJECT_ROOT))
print("Raíz del proyecto:", PROJECT_ROOT)
````

## 1. Funciones con condicionales

````python
def classify_humidity_level(humidity: float) -> str:
    if humidity < 0.40:
        return "Humedad baja"
    elif humidity <= 0.70:
        return "Humedad adecuada"
    else:
        return "Humedad alta"


def classify_temperature_level(temperature: float) -> str:
    if temperature < 20:
        return "Temperatura baja"
    elif temperature <= 35:
        return "Temperatura adecuada"
    else:
        return "Temperatura alta"


def classify_decomposition_speed(percent_remaining: float) -> str:
    if percent_remaining >= 70:
        return "Descomposición lenta"
    elif percent_remaining >= 40:
        return "Descomposición moderada"
    else:
        return "Descomposición rápida"
````

## 2. Probar las funciones

````python
print(classify_humidity_level(0.35))
print(classify_humidity_level(0.55))
print(classify_humidity_level(0.80))

print(classify_temperature_level(15))
print(classify_temperature_level(25))
print(classify_temperature_level(40))

print(classify_decomposition_speed(80))
print(classify_decomposition_speed(50))
print(classify_decomposition_speed(20))
````

## 3. Aplicar condicionales a los resultados del proyecto

````python
import pandas as pd
from src.compost_model import load_sensor_data, load_people_data, run_simulation

sensor_df = load_sensor_data(PROJECT_ROOT / "data" / "datos_compost.xlsx")
people_df = load_people_data(PROJECT_ROOT / "data" / "personas.csv")
results_df = run_simulation(sensor_df, people_df)

results_df["nivel_humedad"] = results_df["humedad"].apply(classify_humidity_level)
results_df["nivel_temperatura"] = results_df["temperatura"].apply(classify_temperature_level)
results_df["velocidad_descomposicion"] = results_df["porcentaje_restante"].apply(classify_decomposition_speed)

display(results_df[["persona", "humedad", "nivel_humedad", "temperatura", "nivel_temperatura", "porcentaje_restante", "velocidad_descomposicion"]].head())
````

## 4. Exportar archivo con clasificaciones

````python
output_dir = PROJECT_ROOT / "outputs"
output_dir.mkdir(exist_ok=True)

output_path = output_dir / "clasificaciones_condicionales_persona_4.csv"
results_df.to_csv(output_path, index=False, encoding="utf-8-sig")

print("Archivo creado:", output_path)
````
