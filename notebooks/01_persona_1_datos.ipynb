# Persona 1 - Datos base

Este notebook corresponde a la participación de la **persona 1**.

Su objetivo es revisar los datos usados para las cinco personas y crear una pequeña documentación en la carpeta `data/`.

````python
from pathlib import Path
import sys

# Detecta la raíz del proyecto aunque abras el notebook desde /notebooks o desde la raíz.
current = Path.cwd()
if (current / "data").exists() and (current / "src").exists():
    PROJECT_ROOT = current
elif (current.parent / "data").exists() and (current.parent / "src").exists():
    PROJECT_ROOT = current.parent
else:
    PROJECT_ROOT = current

sys.path.insert(0, str(PROJECT_ROOT))
print("Raíz del proyecto:", PROJECT_ROOT)
````

## 1. Cargar los datos de personas

Aquí se usa **pandas** para leer el archivo `data/personas.csv`.

````python
import pandas as pd

people_path = PROJECT_ROOT / "data" / "personas.csv"
people_df = pd.read_csv(people_path)
people_df
````

## 2. Validar que existan cinco personas

````python
cantidad_personas = len(people_df)

if cantidad_personas >= 5:
    print(f"Correcto: hay {cantidad_personas} personas registradas.")
else:
    print(f"Faltan personas. Solo hay {cantidad_personas} registros.")
````

## 3. Revisar columnas requeridas

````python
columnas_requeridas = {"persona", "residuo_inicial_kg", "k_descomposicion"}
columnas_actuales = set(people_df.columns)
faltantes = columnas_requeridas - columnas_actuales

if not faltantes:
    print("Correcto: el archivo personas.csv tiene las columnas requeridas.")
else:
    print("Faltan estas columnas:", faltantes)
````

## 4. Crear archivo de descripción de datos

Esta celda crea el archivo `data/descripcion_datos.md`. Ese archivo será la evidencia de la persona 1.

````python
    descripcion = """# Descripción de datos del modelo de compostaje

Este archivo documenta los datos base utilizados para aplicar el modelo de descomposición de residuos orgánicos en cinco personas.

## Variables registradas

- Persona: identifica a cada participante del ejercicio.
- Residuos iniciales: cantidad inicial de residuos orgánicos, expresada en kilogramos.
- Constante k: constante de descomposición usada en la ecuación diferencial.
- Humedad: variable tomada del archivo de sensores.
- Temperatura: variable tomada del archivo de sensores.
- Tiempo: periodo usado para calcular la disminución de residuos.

## Relación con el modelo

Los datos permiten aplicar la ecuación diferencial:

dR/dt = -k * H(t) * T(t) * R(t)

Donde:

- R(t) representa la cantidad de residuos orgánicos.
- k representa la constante de descomposición.
- H(t) representa la humedad.
- T(t) representa la temperatura.

El propósito de estos datos es simular cómo cambia la cantidad de residuos orgánicos según las condiciones de humedad y temperatura.
"""

output_path = PROJECT_ROOT / "data" / "descripcion_datos.md"
output_path.write_text(descripcion, encoding="utf-8")
print("Archivo creado:", output_path)
````
