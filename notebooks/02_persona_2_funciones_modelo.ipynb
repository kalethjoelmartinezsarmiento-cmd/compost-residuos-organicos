## 2. Nueva función propuesta por la persona 2

Esta función calcula el residuo después de avanzar un intervalo de tiempo.

````python
def calculate_residue_after_step(
    current_residue_kg: float,
    derivative: float,
    dt_hours: float
) -> float:
    """
    Calcula el residuo restante después de avanzar un intervalo de tiempo.

    Se usa la aproximación:

        R_nuevo = R_actual + (dR/dt * dt)

    Donde:
    - R_actual: residuo actual en kilogramos.
    - dR/dt: tasa de cambio de los residuos.
    - dt: tiempo transcurrido en horas.
    """
    if current_residue_kg < 0:
        raise ValueError("El residuo actual no puede ser negativo.")

    if dt_hours < 0:
        raise ValueError("El intervalo de tiempo no puede ser negativo.")

    new_residue = current_residue_kg + (derivative * dt_hours)

    return max(new_residue, 0)
````

## 3. Probar la función con un ejemplo

````python
residue_initial = 10
humidity = 0.8
temperature = 20
k = 0.0003
dt_hours = 1

derivative = calculate_derivative(residue_initial, humidity, temperature, k)
residue_after_step = calculate_residue_after_step(residue_initial, derivative, dt_hours)

print("Residuo inicial:", residue_initial)
print("dR/dt:", derivative)
print("Residuo después de una hora:", residue_after_step)
````
