Este notebook corresponde a la participación de la **persona 5**.

Su objetivo es ejecutar pruebas, validar funciones y verificar que el proyecto funcione correctamente.

````python
from pathlib import Path
import sys

# Detecta la raíz del proyecto aunque abras el notebook desde /notebooks o desde la raíz.
current = Path.cwd()
if (current / "data").exists() and (current / "src").exists():
    PROJECT_ROOT = current
elif (current.parent / "data").exists() and (current.parent / "src").exists():
    PROJECT_ROOT = current.parent
else:
    PROJECT_ROOT = current

sys.path.insert(0, str(PROJECT_ROOT))
print("Raíz del proyecto:", PROJECT_ROOT)
````

## 1. Probar funciones principales con `assert`

````python
import pytest
from src.compost_model import calculate_derivative, classify_residue_status, normalize_humidity
import pandas as pd

humidity = normalize_humidity(pd.Series([80, 90, 100]))
assert humidity.tolist() == [0.8, 0.9, 1.0]

derivative = calculate_derivative(residue_kg=10, humidity=0.8, temperature=20, k=0.0003)
assert derivative == pytest.approx(-0.048)

assert classify_residue_status(80) == "Alto"
assert classify_residue_status(50) == "Medio"
assert classify_residue_status(20) == "Bajo"
assert classify_residue_status(10) == "Muy bajo"

print("Pruebas rápidas ejecutadas correctamente.")
````

## 2. Ejecutar la suite de pruebas del proyecto

````python
import subprocess
import sys

result = subprocess.run(
    [sys.executable, "-m", "pytest"],
    cwd=PROJECT_ROOT,
    text=True,
)

if result.returncode == 0:
    print("Todas las pruebas pasaron correctamente.")
else:
    print("Hay pruebas fallando. Revisar la salida anterior.")
````
