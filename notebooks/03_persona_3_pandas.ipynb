# Persona 3 - Procesamiento con pandas

Este notebook corresponde a la participación de la **persona 3**.

Su objetivo es usar **pandas** para cargar, procesar, resumir y exportar datos.

````python
from pathlib import Path
import sys

# Detecta la raíz del proyecto aunque abras el notebook desde /notebooks o desde la raíz.
current = Path.cwd()
if (current / "data").exists() and (current / "src").exists():
    PROJECT_ROOT = current
elif (current.parent / "data").exists() and (current.parent / "src").exists():
    PROJECT_ROOT = current.parent
else:
    PROJECT_ROOT = current

sys.path.insert(0, str(PROJECT_ROOT))
print("Raíz del proyecto:", PROJECT_ROOT)
````

## 1. Cargar archivos con pandas

````python
import pandas as pd

sensor_path = PROJECT_ROOT / "data" / "datos_compost.xlsx"
people_path = PROJECT_ROOT / "data" / "personas.csv"

sensor_df = pd.read_excel(sensor_path)
people_df = pd.read_csv(people_path)

print("Sensores:", sensor_df.shape)
print("Personas:", people_df.shape)
display(sensor_df.head())
display(people_df)
````

## 2. Preparar datos usando funciones del proyecto

````python
from src.compost_model import prepare_sensor_data, validate_people_data

prepared_sensor = prepare_sensor_data(sensor_df)
people = validate_people_data(people_df)

display(prepared_sensor.head())
display(people)
````

## 3. Ejecutar simulación y construir resumen

````python
from src.compost_model import run_simulation, build_summary

results_df = run_simulation(sensor_df, people_df)
summary_df = build_summary(results_df)

display(results_df.head())
display(summary_df)
````

## 4. Usar groupby para analizar resultados por persona

````python
analysis_df = (
    results_df
    .groupby("persona", as_index=False)
    .agg(
        residuo_minimo_kg=("residuo_kg", "min"),
        residuo_maximo_kg=("residuo_kg", "max"),
        promedio_humedad=("humedad", "mean"),
        promedio_temperatura=("temperatura", "mean"),
    )
)

display(analysis_df)
````

## 5. Exportar resultados generados con pandas

````python
output_dir = PROJECT_ROOT / "outputs"
output_dir.mkdir(exist_ok=True)

analysis_path = output_dir / "analisis_pandas_persona_3.csv"
analysis_df.to_csv(analysis_path, index=False, encoding="utf-8-sig")

print("Archivo creado:", analysis_path)
````
